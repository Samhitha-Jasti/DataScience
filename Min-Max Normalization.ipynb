```python
#importing all the necessary packages and iris data set
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#loading iris data set and building the data frame
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
X,y=load_iris(return_X_y=True)
df.head()
df.shape
```

```python
#MIN-MAX Normalization
min_max_Scaler=MinMaxScaler()
min_max_Scaler.fit(X)
# transform the data
X_scaled=min_max_Scaler.transform(X)
# Verify minimum value of all features
X_scaled.min(axis=0)
# Verify maximum value of all features
X_scaled.max(axis=0)
# Manually normalise without using scikit-learn
X_manual_scaled = (X-X.min(axis=0))/(X.max(axis=0)-X.min(axis=0))
# Verify manually VS scikit-learn estimation
print(np.allclose(X_scaled, X_manual_scaled))
```

```python
#Scatter Plot
fig, axes = plt.subplots(1,2)
axes[0].scatter(X[:,0], X[:,1], c=y)
axes[0].set_title("Original data")
axes[1].scatter(X_scaled[:,0], X_scaled[:,1], c=y)
axes[1].set_title("MinMax scaled data")
plt.show()
```

```python
# apply MinMaxScaler for iris data set, [0, 1] for the range
df_n = df.copy()
df_n.iloc[:, [0, 1, 2, 3]] = min_max_Scaler.fit_transform(df_n.iloc[:, [0, 1, 2, 3]])
df_n.head()
```
