```python
#this code helps in creating a dictionary of all words present across all documents
from scipy.sparse import dok_matrix
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer=CountVectorizer()
feature_vector=count_vectorizer.fit(train_data.text)
features=feature_vector.get_feature_names()
#print("Vocabulary: ", count_vectorizer.vocabulary_)
#features.columns=feature_vector.get_feature_names()
print("Total number of features: ",len(features))
#feature_array=np.array(feature_vector)
#df=pd.DataFrame(feature_array)
#count_vectorizer.get_feature_names()
#print(features)
#print(feature_vector)
```

```python
#there are total 92908 unique words or features in the corpus
import random
random.sample(features,15)
```

```python
# converting all documents in dataset to count vectors
train_ds_features=count_vectorizer.transform(train_data.text)
type(train_ds_features)
print(train_ds_features)
train_ds_features.shape
```
